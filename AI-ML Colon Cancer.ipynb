```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

DATA COLLECTION

```python
df = pd.read_csv (r'C:\\Users\\Amogh S Bharadwaj\\Desktop\\colon cancer dataset.csv')
print (df)
```

```python
df.head()
```

DATA PREPROCESSING

```python
df["Class"].unique()
```

```python
df["Class"]=df['Class'].map({'Abnormal':0,'Normal':1})
```

```python
sns.countplot(df['Class'], Label="Class Abnormality",);
```

```python
corr = df.corr()
corr
```

```python
corr.shape
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr);
```

from the above heatmap graph we can see diagonally white line which tells its same diagonally for all types of genes

```python
y = df['Class']
y.head()
```

```python
df['Class'].unique()
```

```python
X = df.drop('Class', axis=1)
X.head()
```

SPLIT THE DATA- TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
df.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

EVALUATING THE MODEL

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)  #input
X_test = ss.transform(X_test)  #squeeze the outputs
```

```python
X_train
```

SCORE MODEL

# MACHINE LEARNING MODELS

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
lr_acc = accuracy_score(y_test, y_pred)
print(lr_acc)
```

```python
results = pd.DataFrame()
results
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'], 'Accuracy':[lr_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
dtc_acc = accuracy_score(y_test, y_pred)
print(dtc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision tree Classifier Method'], 'Accuracy':[dtc_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

# RANDOM FOREST CLASSIFIER 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
rfc_acc = accuracy_score(y_test, y_pred)
print(rfc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Random Forest Classifier Method'], 'Accuracy':[rfc_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

# SUPPORT VECTOR CLASSIFIER 

```python
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
svc_acc = accuracy_score(y_test, y_pred)
print(svc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Support Vector Classifier Method'], 'Accuracy':[svc_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```
